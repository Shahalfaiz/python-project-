```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.txt', sep='\t')
ames = pd.read_csv('ames_housing.txt', sep='\t')
titanic = pd.read_csv('titanic.txt', sep='\t')
batting = pd.read_csv('batting.csv', sep=',')
```

## Problem 1: Iris Dataset

```python
iris.head(8)
```

```python
iris.describe()
```

```python
iris_means = iris.groupby('species').mean()
iris_means
```

```python
iris_colors = ['orange', 'purple', 'pink']
iris_columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
fig = plt.figure(figsize=(8, 6))
for i in range(1, 5):
    plt.subplot(2, 2, i)
    plt.title("Mean " + iris_columns[i - 1])
    plt.bar(iris_means.index, iris_means.iloc[:, i - 1], color=iris_colors, width=0.4, edgecolor="black")
    plt.tight_layout()
plt.show()
```

## Problem 5: Titanic Dataset – Sex Distribution by Class and Outcome

```python
titanic['Female'] = np.where(titanic['Sex'] == 'female', 1, 0)
titanic['Male'] = np.where(titanic['Sex'] == 'male', 1, 0)
titanic.head(8)
```

```python
sex_dist_by_class = titanic[['Pclass', 'Female', 'Male']]
sex_dist_by_class = sex_dist_by_class.groupby('Pclass').sum()
sex_dist_by_class
```

```python
sex_dist_by_outcome = titanic[['Survived', 'Female', 'Male']]
sex_dist_by_outcome = sex_dist_by_outcome.groupby('Survived').sum()
sex_dist_by_outcome
```

## Problem 6: Batting Data – Home Runs

```python
batting.head(5)
```

```python
batting.columns.values
```

```python
total_hr = batting[['teamID', 'yearID', 'HR']]
total_hr = total_hr.groupby(['teamID', 'yearID'], as_index=False).sum()
total_hr = total_hr.sort_values(by='HR', ascending=False)
total_hr.head(10)
```

```python
avg_hr = total_hr[['teamID', 'HR']]
avg_hr = avg_hr.groupby('teamID').mean()
avg_hr = avg_hr.sort_values(by='HR', ascending=False)
avg_hr.head(10)
```

## Problem 7: Batting Data – Batting Averages

```python
batting_avg = batting[['teamID', 'yearID', 'H', 'AB']]
batting_avg = batting_avg.groupby(['teamID', 'yearID']).sum()
batting_avg['BA'] = batting_avg['H'] / batting_avg['AB']
batting_avg = batting_avg.sort_values(by='BA', ascending=False)
batting_avg.head(10)
```

## Problem 8: Batting Data – Cardinals vs. Cubs

```python
stl_batting = batting.loc[(batting['teamID'] == 'SLN') & (batting['yearID'] >= 1900)]
stl_batting = stl_batting.loc[:, ['yearID', 'H', 'AB', 'HR']].groupby('yearID').sum()
stl_batting['BA'] = stl_batting['H'] / stl_batting['AB']
chi_batting = batting.loc[(batting['teamID'] == 'CHN') & (batting['yearID'] >= 1900)]
chi_batting = chi_batting.loc[:, ['yearID', 'H', 'AB', 'HR']].groupby('yearID').sum()
chi_batting['BA'] = chi_batting['H'] / chi_batting['AB']
```

```python
plt.figure(figsize=(12, 4))
years = stl_batting.index
cardinals_ba = stl_batting['BA']
cubs_ba = chi_batting['BA']
cardinals_hr = stl_batting['HR']
cubs_hr = chi_batting['HR']
cardinals_color = 'purple'
cubs_color = 'green'
plt.subplot(1, 2, 1)
plt.plot(years, cardinals_ba, label='Cardinals', color=cardinals_color)
plt.plot(years, cubs_ba, label='Cubs', color=cubs_color)
plt.xlabel("Year")
plt.ylabel("Batting Average")
plt.title("Batting Average By Year")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(years, cardinals_hr, label='Cardinals', color=cardinals_color)
plt.plot(years, cubs_hr, label='Cubs', color=cubs_color)
plt.xlabel("Year")
plt.ylabel("Home Runs")
plt.title("Home Runs by Year")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
proportion_higher_avg = np.mean(stl_batting['BA'] > chi_batting['BA'])
print(round(proportion_higher_avg,4))
```

```python
proportion_more_home_runs = np.mean(stl_batting['HR'] > chi_batting['HR'])
print(round(proportion_more_home_runs,4))
```
